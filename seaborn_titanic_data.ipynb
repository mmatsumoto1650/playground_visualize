```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.stripplot(data=titanic, x='survived', y='age')
```

```python
sns.stripplot(data=titanic, x='survived', y='age',hue ='sex')
```

```python
sns.stripplot(data=titanic, x='survived', y='age',hue ='sex',dodge=True)
```

```python
sns.catplot(data=titanic, kind='strip', x='pclass', y='age', hue='survived', col='sex', dodge=True)
```

```python
sns.violinplot(data=titanic, x='pclass', y='age',hue='survived',split =True)
```

```python
sns.catplot(data=titanic, kind='violin',x='pclass',y='age',hue='survived',col='sex')
```

```python
sns.countplot(x='survived',data=titanic)
```

```python
sns.countplot(x='survived',hue= 'class',data=titanic)
```

```python
sns.barplot(x='class', y='survived', data=titanic)
```

```python
sns.barplot(x='class', y='survived',hue ='sex', data=titanic)
```

```python
fig, axes = plt.subplots(2,2,figsize=(15,10))

sns.countplot(x='survived',data=titanic,ax=axes[0,0])
sns.countplot(x='survived',hue='class',data=titanic,ax=axes[0,1])
sns.barplot(x='class',y='survived',data=titanic,ax=axes[1,0])
sns.barplot(x='class',y='survived',hue='sex',data=titanic,ax=axes[1,1])
```
